```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sns
from sklearn.metrics import mean_squared_error,accuracy_score
```

importing librariessss

```python
data = pd.read_csv('Real estate.csv')
data.head()
data = data.sample(frac=1).reset_index(drop=True)
```

Reading data from csv 'Realestate.csv'













```python
sns.heatmap(data.corr(), annot = True)  # Add a colorbar to the heatmap
plt.title('Heatmap Example')
plt.show()
```

Creating the heatmap for raw data that was provided

```python
data = data.drop(['No','X3 distance to the nearest MRT station','X2 house age'],axis = 1)
print(data.head())
sns.heatmap(data.corr(),annot = True)
```

Creating a heat map after dropping negative corelational values


```python
y_axis = data['Y house price of unit area']
x_axis = data.drop(['Y house price of unit area'],axis = 1)
print(x_axis.head())
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_axis,y_axis,test_size=0.25)
print(x_train.shape)
print(y_train.shape)
```

Preparing and slicing data for linear regression model for sklearn 

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(y_pred.shape)
```

training data with Linear Regression model by scikit_learn

```python

for i in range(100):
    print(f'So The predicted result was {y_pred[i]} and actual result was {y_test.values[i]}')
```

```python

print(mean_squared_error(y_test,y_pred)/100)

```
